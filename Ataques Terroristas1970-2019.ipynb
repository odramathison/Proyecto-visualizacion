```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
df = pd.read_excel('globalterrorismdb_0221dist.xlsx')
```

```python
df_new= df.copy()
```

```python
df_new.dropna(axis=1, how = 'all')
```

```python
print(df_new.columns)
```

```python
df_new.rename(columns={'iyear':'Year', 'imonth':'Month', 'iday':'Day', 'country_txt':'Country', 'region_txt':'Region',
                       'attacktype1_txt':'AttackType','target1':'Target', 'nkill':'Killed', 'nwound':'Wounded', 'summary':'Summary',
                      'gname':'Group','targtype1_txt':'Target_type', 'weaptype1_txt':'Weapon_type', 'motive':'Motive'},inplace = True)
```

```python
print(df_new.columns)
```

```python
df_new=df_new[['Year', 'Month','Day', 'Country','Region', 'city' , 'latitude', 'longitude', 'AttackType', 'Killed', 
               'Wounded', 'Target','Summary', 'Group', 'Target_type', 'Weapon_type', 'Motive']]
```

```python
df_new['casualities']= df_new['Killed']+ df_new['Wounded']
```

```python
df_new.head(5)
```

```python
df_new.isnull().sum()
```

## Basic Analysis

```python
print('Pais con mayor cantidad de ataques terroristas:',df_new['Country'].value_counts().index[0])
print('Region con mayor cantidad de ataques terroristas:',df_new['Region'].value_counts().index[0])
print('Region con mayor cantidad de tipo de arma en ataques terroristas:',df_new['Weapon_type'].value_counts().index[0])
print('Region con mayor cantidad de personas muertas en ataques terroristas:',df_new['Killed'].max(), 'que ocurrio en:', 
      df_new.loc[df_new['Killed'].idxmax()].Country)
```

## Number of Terrorist Activities vs Year

```python
plt.subplots(figsize=(18,8))
sns.countplot('Year', data=df_new, palette = 'GnBu', edgecolor=sns.color_palette('dark',7))
plt.xticks(rotation=60)
plt.title('Number of Terrorist Activities Each Year', size=18)
plt.show()
```

## What are the attacking methods used?

```python
plt.subplots(figsize= (18,8))
sns.countplot('AttackType', data= df_new, palette = 'gist_rainbow_r', order = df_new['AttackType'].value_counts().index)
plt.xticks(rotation=60)
plt.title('Attacking Methods by Terrorists', size=18)
plt.show()
```

## Who are the targets?

```python
plt.subplots(figsize= (18,8))
sns.countplot('Target_type', data= df_new, palette = 'gist_rainbow_r', order = df_new['Target_type'].value_counts().index)
plt.xticks(rotation=65)
plt.title('Favorite Targets', size=18)
plt.show()
```

## Hot zones of terrorism

```python
plt.subplots(figsize= (18,8))
sns.countplot('Region', data= df_new, palette = 'hot', edgecolor= sns.color_palette('dark',7), order = df_new['Region'].value_counts().index)
plt.xticks(rotation=65)
plt.title('Number of Terrorist Activities by Region', size=18)
plt.show()
```

## Attack Type vs Region

```python
pd.crosstab(df_new.Region, df_new.AttackType).plot.barh(stacked=True, width=1, color=sns.color_palette('ocean', 9))
fig= plt.gcf()
fig.set_size_inches(18,8)
plt.title('Attack Type vr Region', size=18)
plt.show()
```

## Terrorism by Country

```python
plt.subplots(figsize= (18,8))
sns.barplot(df_new['Country'].value_counts()[:15].index, df_new['Country'].value_counts()[:15].values, palette = 'plasma_r')
plt.xticks(rotation=65)
plt.title('Top Affected Countries', size=18)
plt.show()
```

## Attacks vs Killed

```python
coun_terror = df_new['Country'].value_counts()[:15].to_frame()
coun_terror.columns = ['Attacks']
coun_kill = df_new.groupby('Country')['Killed'].sum().to_frame()
coun_terror.merge(coun_kill, left_index= True, right_index = True, how= 'left').plot.bar(width=0.8)
plt.title('Attacks vs Killed', size=18)
fig= plt.gcf()
fig.set_size_inches (18,8)
plt.show()
```

## Most Notorious Groups

```python
sns.barplot(df_new['Group'].value_counts()[1:15].values, df_new['Group'].value_counts()[1:15].index, palette= ('CMRmap'))
plt.title('Terrorist Groups with Highest Terror Attacks', size=18)
plt.xticks(rotation=90)
fig= plt.gcf()
fig.set_size_inches (18,8)
plt.show()

```

## Activity of Top Terrorist Groups

```python
top_groups_10= df_new[df_new['Group'].isin(df_new['Group'].value_counts()[1:11].index)]
pd.crosstab(top_groups_10.Year, top_groups_10.Group).plot(color= sns.color_palette('magma',10))
fig= plt.gcf()
fig.set_size_inches (18,8)
plt.show()
```

## Most Notorious Groups in India and Favorite Attack Types

```python
terror_india= df_new[df_new['Country']=='India']
f, ax= plt.subplots(1,2, figsize=(25,15))
ind_groups=terror_india['Group'].value_counts()[1:11].index
ind_groups=terror_india[terror_india['Group'].isin(ind_groups)]
sns.countplot(y='Group',data=ind_groups, ax=ax[0])
ax[0].set_title('Top Terrorist Groups', size=(20))
sns.countplot(y='AttackType',data=terror_india, ax=ax[1])
ax[1].set_title('Favorite Attack Types', size=(20))
plt.subplots_adjust(hspace=0.5, wspace=0.8)
ax[0].tick_params(labelsize=20)
ax[1].tick_params(labelsize=20)
plt.show()
```
